# Problem 1


**Imports**

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

**First Function: f(x,y) = y^2 - x^2**

```python
def f(x, y):
    return y**2 - x**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(X,Y,Z)

scatterplot = ax.scatter3D(X, Y, Z)
```

**Contour Plot for f(x,y)**

Here you can see the saddle point (0,0) marked in red

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(30,50)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.contour(X,Y,Z)
ax.plot([0], [0], [0], markerfacecolor='red', markeredgecolor='red', marker='o', markersize=6, alpha=0.6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

**Second Function: g(x,y) = 2x^3 - 6xy + y^2 + 4y**

```python
def g(x, y):
  # return 2*x**3 - 6*x*y + y**2 + 4*y
  return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = g(X, Y)
```

**Scatter Plot for g(x,y)** 

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(X,Y,Z)

scatterplot = ax.scatter3D(X, Y, Z)
```

**Contour Plot for g(x,y)**

Here you can see a local min at (0,0,0) marked in red

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(67, 75)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.contour(X,Y,Z)
ax.plot([0], [0], [0], markerfacecolor='red', markeredgecolor='red', marker='o', markersize=6, alpha=0.6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

**Third Function h(x) = xy - 3x -x^2 - y^2**

```python
def h(x, y):
  return x*y - 3*x - 3*y - x**2 - y**2

x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)

X, Y = np.meshgrid(x, y)
Z = h(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour(X,Y,Z, colors='red')

scatterplot = ax.scatter3D(X, Y, Z)
```

Here you can see the local max at (-3,-3,9)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.contour(X,Y,Z)
ax.view_init(30,30)
ax.plot([-3], [-3], [9], markerfacecolor='red', markeredgecolor='red', marker='o', markersize=6, alpha=0.6)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
