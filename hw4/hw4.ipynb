### Artificial Intelligence (AI)

AI is an umbrella term for a vast array of subcategories of Computer Science that deal with the theory and development of systems which aim to accomplish goals that could typically only be done by a human. 

### Machine Learning (ML)

A subcategory of AI, Machine Learning focuses on creating statistical models that aim to accomplish a goal without starting off with a set of rules. In clearer terms, where a normal program is a set of rules that deal with input and release some output, machine learning algorithms take in input and then output the rules. 

### Deep Learning (DL) 

Deep Learning is a subset of Machine Learning that focuses on 'unsupervised' learning. In Deep Neural Networks, there is a hierarchical system of artificial neural networks which carry out the maching learning process. A huge benefit of Deep Neural networks is that they can work with unlabeled and unstructured data, which typical machine learning models struggle with.

### Gradient Descent

Gradient descent is an algorithm which is used to find the minimum, or local minimum of a function. This is extremely useful in general, but especially so for the case of machine learning. 

![alt text](https://ml-cheatsheet.readthedocs.io/en/latest/_images/gradient_descent_demystified.png) (Source: https://ml-cheatsheet.readthedocs.io/en/latest/gradient_descent.html)

You can sort of get the idea from the above picture. In essence, we start at the top and work our way down until we cant go down any further, and at that point we've reached a local minimum.

#### Learning Rate

You may be asking, "how do we know how far to move as we step along looking for our local minimum?". Well, the distance of the steps we take is what we call learning rate. If our learning rate is high, it'll take us less steps to find the minimum, but we risk overshooting and moving way past the lowest point. On the other hand, with a very low learning rate you can much more accurately find the minimum, but the number of calculations needed grows by a lot and it can be costly on performance. 

#### Loss Functions

Well, how can we evaluate how well our models are learning? This is where loss functions come in. The slope of the curve of the loss function tells us how our parameters should be updated in order to improve accuracy.

#### Layers

The layers of a neural network are:

1) **Input Layer**: This is where the data that you're using to train your model, AKA your input, will be. All the neurons in this layer represent a unique feature in your dataset. For the dog and cat example, these features could be hair color, face shape, etc. 

2) **Hidden Layer**: This is where the activation functions are applied. Most of the time a network will have more than one hidden layer, where each successive layer works on the output from the previous layer. 

3) **Output Layer**: The output layer receives input from the final hidden layer and returns the prediction of your model. It can also apply an activation function if you please.
